```python
import matplotlib.pyplot as plt
```

```python
with open('bubble.txt', 'r') as f:
    lines = f.readlines()

    bubRandX = [i for i in range(len(lines))]
    bubRandY = [(int)(line.split()[1]) for line in lines]
```

```python
with open('insertion.txt', 'r') as f:
    lines = f.readlines()

    insRandX = [i for i in range(len(lines))]
    insRandY = [(int)(line.split()[1]) for line in lines]
```

```python
with open('selection.txt', 'r') as f:
    lines = f.readlines()

    selRandX = [i for i in range(len(lines))]
    selRandY = [(int)(line.split()[1]) for line in lines]
```

# Building graphics

```python
plt.xlabel('Element count (1000 - 10^5)', color = 'gray')
plt.ylabel('Time (ms)', color = 'gray')

plt.title('Quadratic sortings comparison ', fontsize = 15)

plt.plot(bubRandX, bubRandY)
plt.plot(selRandX, selRandY)
plt.plot(insRandX, insRandY)

plt.legend(['Bubble', 'Selection', 
            'Insertion'])

plt.xticks([])  # removing numbers on x
plt.show()
```

```python
plt.xlabel('Element count (1000 - 10^7)', color = 'gray')
plt.ylabel('Time (ms)', color = 'gray')

plt.title('Qsort with different pivots', fontsize = 15)

plt.plot(qRandX, qRandY)
plt.plot(medRandX, medRandY)
plt.plot(med3RandX, med3RandY)
plt.plot(centRandX, centRandY)

plt.legend(['Random', 'Median of medians', 
            'Median of 3', 'Center'])

plt.xticks([])  # removing numbers on x
plt.show()
```

```python
plt.xlabel('Element count (1000 - 10^7)', color = 'gray')
plt.ylabel('Time (ms)', color = 'gray')

plt.title('"Fast" sortings comparison', fontsize = 15)

plt.plot(qRandX, qRandY)
plt.plot(med3RandX, med3RandY)
plt.plot(centRandX, centRandY)
plt.plot(mergeRandX, mergeRandY)
plt.plot(lsdRandX, lsdRandY)

plt.legend(['QSort with random pivot',
            'QSort with median of 3 pivot', 'QSort with center pivot',
            'Merge sort', 'LSD'])

plt.xticks([])  # removing numbers on x
plt.show()
```
